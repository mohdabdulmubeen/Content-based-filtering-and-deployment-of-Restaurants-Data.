```python
import ast
```

```python
weeks = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
```

```python
# conversion of string to dictionary and change order of days as in weeks list 

for i,j in enumerate(df['Timing']):
    week_times = ast.literal_eval(j)
    if "Description" in week_times:
        continue
    new_week = dict()
    new_week.update((week, week_times[week]) for week in weeks)
    print(new_week)
    df['Timing'][i] = new_week
```

```python
shop_input= input('Enter the Restaurant:')
row_selected = df.loc[df["Name"] == shop_input]
if len(row_selected) == 0:
    print("No Shop found")
else:
    time_input = input('Enter Time:') 
    day_input = input('Enter the day:')
    timing_cell = row_selected['Timing'].item()


    if 'Description' in timing_cell:
        print('shop is temporarily closed')

    else:
        if timing_cell[day_input].strip() == 'Closed':
            print('shop is closed')

        elif timing_cell[day_input].strip() == 'Open 24 hours':
            print('the shop is open')

        else:
            start, end = timing_cell[day_input].split('–')
            time_input_hour = int(time_input.split(':')[0])
            start_hour = int(start.split(':')[0])
            end_hour = int(end.split(':')[0])
            if start_hour <= time_input_hour <= end_hour:
                print('the shop is open')
            else:
                print('shop is closed')
```
